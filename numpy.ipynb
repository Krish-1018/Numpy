## random in numpy

### np.random.rand()
it generates random number its default range is b/w 0 and 1 but close to 0

```python
p = np.random.rand(4)
print(p)
```

```python
a = np.random.rand(4,4)
print(a)
```

### np.random.randn()
it generates numbers but more likely to be negative and also try to maintain the space

```python
a = np.random.randn(2)
print(a)
```

```python
c = np.random.randn(4,4)
print(c)
```

### broadcasting

```python
arr = np.arange(0,10)
print(arr)
```

```python
slice_arr = arr[0:5]
print(slice_arr)
```

```python
slice_arr[0] = 99
print(slice_arr)
```

```python
slice_arr[:] = 99
print(slice_arr)
```

```python
arr
```

```python
# copy()
arr_copy = arr.copy()
print(arr_copy)
```

```python
arr_copy[:4] = 45
print(arr_copy)
```
